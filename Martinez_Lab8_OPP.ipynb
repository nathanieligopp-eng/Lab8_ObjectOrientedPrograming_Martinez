```python
class ATM:
    def __init__(self, balance):
        self.balance = balance

    def withdraw(self, amount):
        if amount > self.balance:
            return "Insufficient balance"
        self.balance -= amount
        return f"Withdrawn {amount}, Remaining balance: {self.balance}"

atm = ATM(1000)
print(atm.withdraw(200))
```

```python
# Try It: Multiple Inheritance and MRO
class Base:
    def show(self):
        return "Message from Base Class"

class Parent1(Base):
    def show(self):
          return "Message from Parent1"

class Parent2(Base):
     def show(self):
         return "Message from Parent2"

# Class D inherits from both Parent1 and Parent2
class ChildD(Parent1, Parent2):
    pass

# Create instance and test behavior
d_obj = ChildD()
print(f"Executing show(): {d_obj.show()}")

# Display the MRO (the order Python searches for methods)
print("\nMethod Resolution Order (MRO):")
for cls in ChildD.mro():
    print(cls)
```

```python
# Step 1: Define multiple classes
class A:
    def show(self):
        return "Class A"

class B(A):
    def show(self):
        return "Class B"

class C(A):
    def show(self):
        return "Class C"

class D(B, C):
    pass # Inherits show() based on MRO

# Step 2: Create an object of class D and call the method
obj = D()
print(obj.show()) # Output depends on MRO
```

```python
# Step 1: Import the ABC (Abstract Base Class) logic
from abc import ABC, abstractmethod

# Step 1: Define an abstract class
class Appliance(ABC):
    @abstractmethod
    def turn_on(self):
        pass

# Step 2: Define a subclass implementing the abstract method
class WashingMachine(Appliance):
    def turn_on(self):
        return "Washing machine is now running."

# Step 3: Create an object of the subclass and call the method
wm = WashingMachine()
print(wm.turn_on())
```

```python
# Try it: Vector Addition and Subtraction
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # overload '+'
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

# Create instances
v1 = Vector(10, 5)
v2 = Vector(3, 2)

# Test the Operators
print(f"Addition: {v1 + v2}")
print(f"Subtraction: {v1 - v2}")
```

```python
# Step 1: Define a base class
class Shape:
    def area(self):
        return "Calculating area..."

# Step 2: Defin subclass with their implementations
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

# Step 3: Create objects and call their methods
circle = Circle(5)
rectangle = Rectangle(4, 6)

print(f"Circle area: {circle.area()}")
print(f"Rectangle area: {rectangle.area()}")
```
